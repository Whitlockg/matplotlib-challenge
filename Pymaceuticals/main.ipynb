```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
Mouse_df=pd.read_csv(mouse_drug_data_to_load)
Trial_df=pd.read_csv(clinical_trial_data_to_load)

Mouse_df.head()
```

```python
Trial_df.head()
```

```python
# Combine the data into a single dataset
combined_df = pd.merge(Trial_df, Mouse_df,how='outer', on="Mouse ID")
```

```python
# Display the data table for preview
combined_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tv = combined_df.groupby(["Drug","Timepoint"]).mean()
mean_tv.reset_index(level = None, inplace = True)

# Convert toDataFrame
mean_df = pd.DataFrame(mean_tv)

# Preview DataFrame 
mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
combined_ste = combined_df.groupby(["Drug","Timepoint"]).sem()
combined_ste.reset_index(level = None, inplace = True)

# Convert to DataFrame
combined_ste_df = pd.DataFrame(combined_ste)

# Preview DataFrame 
combined_ste_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatsem_df= combined_ste_df.pivot(index ="Timepoint", columns = 'Drug', values = "Tumor Volume (mm3)")
reformatsem_df.reset_index(level = None, inplace = True)

#Preview the formatted table
reformatsem_df.head()
```

```python
finalsem_df = reformatsem_df[["Timepoint","Capomulin", "Infubinol", "Ketapril", "Placebo"]]
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatm_df= mean_df.pivot(index ="Timepoint", columns = 'Drug', values = "Tumor Volume (mm3)")
reformatm_df.reset_index(level = None, inplace = True)

#Preview the formatted table
reformatm_df.head()
```

```python
finaltr_df = reformatm_df[["Timepoint","Capomulin", "Infubinol", "Ketapril", "Placebo"]]
finaltr_df.head()
```

```python
plt.figure(figsize = (10, 5))

X=finaltr_df['Timepoint']
Y1=finaltr_df['Capomulin']
Y2=finaltr_df['Infubinol']
Y3=finaltr_df['Ketapril']
Y4=finaltr_df['Placebo']
Y_error1=finalsem_df['Capomulin']
Y_error2=finalsem_df['Infubinol']
Y_error3=finalsem_df['Ketapril']
Y_error4=finalsem_df['Placebo']

Capomulin=plt.errorbar(x=X, y=Y1, yerr=Y_error1, linestyle="--", fmt='o', label='Capomulin')
Infubinol=plt.errorbar(x=X, y=Y2, yerr=Y_error2, linestyle="--", fmt='o', label='Infubinol')
Ketapril=plt.errorbar(x=X, y=Y3, yerr=Y_error3, linestyle="--", fmt='o', label='Ketapril')
Placebo=plt.errorbar(x=X, y=Y4, yerr=Y_error4, linestyle="--", fmt='o', label='Placebo')

plt.ylabel('Tumor Volume(mm3)')
plt.xlabel('Time (Days)')
plt.title('Tumor Response to Treatment')
plt.grid()
plt.legend(loc='upper left')       
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met = combined_df.groupby(["Drug","Timepoint"]).mean()
mean_met.reset_index(level = None, inplace = True)

# Convert toDataFrame
mean_met_df = pd.DataFrame(mean_met)

# Preview DataFrame 
mean_met_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
combined_metsem = combined_df.groupby(["Drug","Timepoint"]).sem()
combined_metsem.reset_index(level = None, inplace = True)

# Convert to DataFrame
combined_metsem_df = pd.DataFrame(combined_metsem)

# Preview DataFrame 
combined_metsem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatmetsem_df= combined_metsem_df.pivot(index ="Timepoint", columns = 'Drug', values = "Tumor Volume (mm3)")
reformatmetsem_df.reset_index(level = None, inplace = True)

#Preview the formatted table
reformatmetsem_df.head()
```

```python
finalmetsem_df = reformatmetsem_df[["Timepoint","Capomulin", "Infubinol", "Ketapril", "Placebo"]]
finalmetsem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatmean_met_df= mean_met_df.pivot(index ="Timepoint", columns = 'Drug', values = "Tumor Volume (mm3)")
reformatmean_met_df.reset_index(level = None, inplace = True)

#Preview the formatted table
reformatmean_met_df.head()
```

```python
finalmean_met_df = reformatmean_met_df[["Timepoint","Capomulin", "Infubinol", "Ketapril", "Placebo"]]
finalmean_met_df.head()
```

```python
plt.figure(figsize = (10, 5))

X=finalmean_met_df['Timepoint']
Y1=finalmean_met_df['Capomulin']
Y2=finalmean_met_df['Infubinol']
Y3=finalmean_met_df['Ketapril']
Y4=finalmean_met_df['Placebo']
Y_error1=finalmetsem_df['Capomulin']
Y_error2=finalmetsem_df['Infubinol']
Y_error3=finalmetsem_df['Ketapril']
Y_error4=finalmetsem_df['Placebo']

Capomulin=plt.errorbar(x=X, y=Y1, yerr=Y_error1, linestyle="--", fmt='o', label='Capomulin')
Infubinol=plt.errorbar(x=X, y=Y2, yerr=Y_error2, linestyle="--", fmt='o', label='Infubinol')
Ketapril=plt.errorbar(x=X, y=Y3, yerr=Y_error3, linestyle="--", fmt='o', label='Ketapril')
Placebo=plt.errorbar(x=X, y=Y4, yerr=Y_error4, linestyle="--", fmt='o', label='Placebo')

plt.ylabel('Met. Sites')
plt.xlabel('Treatment Duration (Days)')
plt.title('Metastatic Spread During Treatment')
plt.grid()
plt.legend(loc='upper left')       
plt.show()
```
