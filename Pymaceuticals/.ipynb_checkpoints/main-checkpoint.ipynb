```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
Mouse_df=pd.read_csv(mouse_drug_data_to_load)
Trial_df=pd.read_csv(clinical_trial_data_to_load)

Mouse_df.head()
```

```python
Trial_df.head()
```

```python
# Combine the data into a single dataset
combined_df = pd.merge(Trial_df, Mouse_df,how='outer', on="Mouse ID")
```

```python
# Display the data table for preview
combined_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tv = combined_df.groupby(["Drug","Timepoint"]).mean()
mean_tv.reset_index(level = None, inplace = True)

# Convert toDataFrame
mean_df = pd.DataFrame(mean_tv)

# Preview DataFrame 
mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
combined_ste = combined_df.groupby(["Drug","Timepoint"]).sem()
combined_ste.reset_index(level = None, inplace = True)

# Convert to DataFrame
combined_ste_df = pd.DataFrame(combined_ste)

# Preview DataFrame 
combined_ste_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatsem_df= combined_ste_df.pivot(index ="Timepoint", columns = 'Drug', values = "Tumor Volume (mm3)")
reformatsem_df.reset_index(level = None, inplace = True)

#Preview the formatted table
reformatsem_df.head()
```

```python
finalsem_df = reformatsem_df[["Timepoint","Capomulin", "Infubinol", "Ketapril", "Placebo"]]
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatm_df= mean_df.pivot(index ="Timepoint", columns = 'Drug', values = "Tumor Volume (mm3)")
reformatm_df.reset_index(level = None, inplace = True)

#Preview the formatted table
reformatm_df.head()
```

```python
finaltr_df = reformatm_df[["Timepoint","Capomulin", "Infubinol", "Ketapril", "Placebo"]]
finaltr_df.head()
```

```python
plt.figure(figsize = (10, 5))

X=finaltr_df['Timepoint']
Y1=finaltr_df['Capomulin']
Y2=finaltr_df['Infubinol']
Y3=finaltr_df['Ketapril']
Y4=finaltr_df['Placebo']
Y_error1=finalsem_df['Capomulin']
Y_error2=finalsem_df['Infubinol']
Y_error3=finalsem_df['Ketapril']
Y_error4=finalsem_df['Placebo']

Capomulin=plt.errorbar(x=X, y=Y1, yerr=Y_error1, linestyle="--", fmt='o', label='Capomulin')
Infubinol=plt.errorbar(x=X, y=Y2, yerr=Y_error2, linestyle="--", fmt='o', label='Infubinol')
Ketapril=plt.errorbar(x=X, y=Y3, yerr=Y_error3, linestyle="--", fmt='o', label='Ketapril')
Placebo=plt.errorbar(x=X, y=Y4, yerr=Y_error4, linestyle="--", fmt='o', label='Placebo')

plt.ylabel('Tumor Volume(mm3)')
plt.xlabel('Time (Days)')
plt.title('Tumor Response to Treatment')
plt.grid()
plt.legend(loc='upper left')       
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met = combined_df.groupby(["Drug","Timepoint"]).mean()
mean_met.reset_index(level = None, inplace = True)

# Convert toDataFrame
mean_met_df = pd.DataFrame(mean_met)

# Preview DataFrame 
mean_met_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
combined_metsem = combined_df.groupby(["Drug","Timepoint"]).sem()
combined_metsem.reset_index(level = None, inplace = True)

# Convert to DataFrame
combined_metsem_df = pd.DataFrame(combined_metsem)

# Preview DataFrame 
combined_metsem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatmetsem_df= combined_metsem_df.pivot(index ="Timepoint", columns = 'Drug', values = "Metastatic Sites")
reformatmetsem_df.reset_index(level = None, inplace = True)

#Preview the formatted table
reformatmetsem_df.head()
```

```python
finalmetsem_df = reformatmetsem_df[["Timepoint","Capomulin", "Infubinol", "Ketapril", "Placebo"]]
finalmetsem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatmean_met_df= mean_met_df.pivot(index ="Timepoint", columns = 'Drug', values = "Metastatic Sites")
reformatmean_met_df.reset_index(level = None, inplace = True)

#Preview the formatted table
reformatmean_met_df.head()
```

```python
finalmean_met_df = reformatmean_met_df[["Timepoint","Capomulin", "Infubinol", "Ketapril", "Placebo"]]
finalmean_met_df.head()
```

```python
plt.figure(figsize = (10, 5))

X=finalmean_met_df['Timepoint']
Y1=finalmean_met_df['Capomulin']
Y2=finalmean_met_df['Infubinol']
Y3=finalmean_met_df['Ketapril']
Y4=finalmean_met_df['Placebo']
Y_error1=finalmetsem_df['Capomulin']
Y_error2=finalmetsem_df['Infubinol']
Y_error3=finalmetsem_df['Ketapril']
Y_error4=finalmetsem_df['Placebo']

Capomulin=plt.errorbar(x=X, y=Y1, yerr=Y_error1, linestyle="--", fmt='o', label='Capomulin')
Infubinol=plt.errorbar(x=X, y=Y2, yerr=Y_error2, linestyle="--", fmt='o', label='Infubinol')
Ketapril=plt.errorbar(x=X, y=Y3, yerr=Y_error3, linestyle="--", fmt='o', label='Ketapril')
Placebo=plt.errorbar(x=X, y=Y4, yerr=Y_error4, linestyle="--", fmt='o', label='Placebo')

plt.ylabel('Met. Sites')
plt.xlabel('Treatment Duration (Days)')
plt.title('Metastatic Spread During Treatment')
plt.grid()
plt.legend(loc='upper left')       
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count=combined_df.groupby(["Drug","Timepoint"]).count()

# Convert to DataFrame
mice_count_df=pd.DataFrame(mice_count["Mouse ID"])
mice_count_df.reset_index(inplace=True)

#Display dataframe
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatmice_count_df= mice_count_df.pivot(index ="Timepoint", columns = 'Drug', values = "Mouse ID")
reformatmice_count_df.reset_index(level = None, inplace = True)

#Preview the formatted table
reformatmice_count_df.head()
```

```python
reformatmice_count_df = reformatmice_count_df[["Timepoint","Capomulin", "Infubinol", "Ketapril", "Placebo"]]
reformatmice_count_df.head()
```

```python
final_mice_count_df = reformatmice_count_df.astype(float)
final_mice_count_df["Capomulin_Survival_Rate"]=final_mice_count_df["Capomulin"]/final_mice_count_df["Capomulin"].iloc[0] * 100
final_mice_count_df["Infubinol_Survival_Rate"]=final_mice_count_df["Infubinol"]/final_mice_count_df["Infubinol"].iloc[0] * 100
final_mice_count_df["Ketapril_Survival_Rate"]=final_mice_count_df["Ketapril"]/final_mice_count_df["Ketapril"].iloc[0] * 100
final_mice_count_df["Placebo_Survival_Rate"]=final_mice_count_df["Placebo"]/final_mice_count_df["Placebo"].iloc[0] * 100
final_mice_count_df
```

```python
plt.figure(figsize = (10, 5))

X=final_mice_count_df['Timepoint']
Y1=final_mice_count_df["Capomulin_Survival_Rate"]
Y2=final_mice_count_df["Infubinol_Survival_Rate"]
Y3=final_mice_count_df["Ketapril_Survival_Rate"]
Y4=final_mice_count_df["Placebo_Survival_Rate"]

Capomulin=plt.errorbar(x=X, y=Y1, linestyle="--", fmt='o', label='Capomulin')
Infubinol=plt.errorbar(x=X, y=Y2, linestyle="--", fmt='o', label='Infubinol')
Ketapril=plt.errorbar(x=X, y=Y3, linestyle="--", fmt='o', label='Ketapril')
Placebo=plt.errorbar(x=X, y=Y4, linestyle="--", fmt='o', label='Placebo')

plt.ylabel('Survival During Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.title('Survival Rate (%)')
plt.grid()
plt.legend(loc='upper left')       
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for Capomulin drug
pc_Capomulin=(finaltr_df["Capomulin"].iloc[9]-finaltr_df["Capomulin"].iloc[0])/finaltr_df["Capomulin"].iloc[0]*100
pc_Infubinol=(finaltr_df["Infubinol"].iloc[9]-finaltr_df["Infubinol"].iloc[0])/finaltr_df["Infubinol"].iloc[0]*100
pc_Ketapril=(finaltr_df["Ketapril"].iloc[9]-finaltr_df["Ketapril"].iloc[0])/finaltr_df["Ketapril"].iloc[0]*100
pc_Placebo=(finaltr_df["Placebo"].iloc[9]-finaltr_df["Placebo"].iloc[0])/finaltr_df["Placebo"].iloc[0]*100

# Display the data to confirm
print(pc_Capomulin)
print(pc_Infubinol)
print(pc_Ketapril)
print(pc_Placebo)
```

```python
# Store all Relevant Percent Changes into a Tuple
pc_tuple= {'Capomulin': pc_Capomulin, 'Infubinol': pc_Infubinol, 'Ketapril': pc_Ketapril, 'Placebo': pc_Placebo}
pc_series = pd.Series(pc_tuple)
pc_series
```

```python
drug_index=pc_series.keys()
drug_index
```

```python
# Splice the data between passing and failing drugs
summary_bar_graph = plt.subplot()
x_axis = np.arange(0, len(drug_index))

# Orient widths. Add labels, tick marks, etc. 
ticks = []
for x in x_axis:
    ticks.append(x + 1)
plt.xticks(ticks, drug_index)
colors = []

for p in pc_series:
    if p >= 0 :
        colors.append('r')
    else:
        colors.append('g')
        
#Plot       
pc_bar_graph = summary_bar_graph.bar(x_axis, pc_series, color=colors, align="edge")

plt.title("Tumor Change Over 45-Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.xlabel("Drug", labelpad=15)
plt.xlim(-0.25, len(drug_index))
plt.ylim(-30, max(pc_series) + 20)
plt.grid()

# Create labels
count = 0

#loops through tumor change data and places the value in the correct position based on + or - values
for r in pc_series:
    if r < 0:
        y_coor = -5
    else:
        y_coor = 5
    plt.text(count, y_coor, str(round(r, 1)) + '%', color = 'white')
    count += 1
    
    plt.subplots_adjust(bottom= 0.2, top = 0.98)

plt.show()
```
